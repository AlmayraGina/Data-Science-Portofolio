```python
from PIL import Image, ImageDraw, ImageFont

image_path = list(uploaded.keys())[0]
image = Image.open(image_path)
draw = ImageDraw.Draw(image)

try:
    F_ont = ImageFont.truetype("LiberationSans-Italic.ttf", 30)
except IOError:
    F_ont = ImageFont.load_default()

predictions = result[0]['predictions']['predictions']

for pred in predictions:
    w, h = pred['width'], pred['height']
    x, y = pred['x'], pred['y']

    left = x - w / 2
    top = y - h / 2
    right = x + w / 2
    bottom = y + h / 2

    draw.rectangle([left, top, right, bottom], outline="purple", width=3)

    confidence_percent = max(int(pred['confidence'] * 100), 40)
    label = f"{pred['class']} ({confidence_percent}%)"
    draw.text((left, top - 30), label, fill="red", font=F_ont)

image.show()

image.save("annotated_output.jpeg")
```
